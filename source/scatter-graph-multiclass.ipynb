#### Load the dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('../dataset/preprocessed-sam-dataset.csv', sep='|',
                 dtype = {'CZ': 'float32', 'FZ': 'float32', 'Fp1': 'float32', 'F3': 'float32',
                          'FC1': 'float32', 'FC5': 'float32', 'FT9': 'float32', 'T7': 'float32',
                          'CP5': 'float32', 'P3': 'float32', 'P7': 'float32', 'PO9': 'float32',
                          'PZ': 'float32', 'O2': 'float32', 'P4': 'float32', 'CP6': 'float32',
                          'FT10': 'float32', 'FC6': 'float32', 'F8': 'float32', 'Fp2': 'float32',
                          'Scale': 'int8'})

colors_list = ['green', 'black', 'blue', 'red', 'yellow', 'cyan', 'magenta', '#81fcd5', '#d47e1c', '#9ad41c', '#b0b7b8']
```

#### Display the dataset size.

```python
df.shape
```

#### Generate the scatter plot using the CZ and FZ features to evaluate the linearity or non-linearity of the data.
- The data is non-linear.

```python
plt.figure(figsize = (400, 400), dpi = 80)
for count in range(0, 11) :
    df_class = df.loc[df['Scale'] == count]
    plt.scatter(df_class.iloc[:, 0], df_class.iloc[:, 1], color = colors_list[count], marker = 'o',
                label = 'class {}'.format(count))

plt.xlabel('CZ')
plt.ylabel('FZ')
plt.legend(loc = 'upper left')
plt.gcf().set_size_inches((10, 10)) 
plt.show()
```

#### Generate the graph of parallel coordinates.

```python
fig, ax = plt.subplots(figsize=(20, 5), dpi=100)
_ = pd.plotting.parallel_coordinates(df, 'Scale', sort_labels=True, axvlines = True, alpha = 0.2, ax = ax)
_ = ax.set_title('Basic parallel coordinate plot')
```

```python
fig2, ax2 = plt.subplots(figsize = (6, 5), dpi = 100)
_ = pd.plotting.parallel_coordinates(df, 'Scale', ['CZ', 'FZ'], 
                                     color = ['green', 'black', 'blue', 'red', 'yellow', 'cyan', 
                                              'magenta', '#81fcd5', '#d47e1c', '#9ad41c', '#b0b7b8'],
                                     sort_labels=True, axvlines = True, alpha = 0.2, ax = ax2)
_ = ax2.set_title('Basic parallel coordinate plot')
```
